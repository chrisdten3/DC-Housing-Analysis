```python
import numpy as np
import pandas as pd
import seaborn as sns
houses = pd.read_csv("./DC_Properties.csv",index_col="Unnamed: 0")
houses.head()
```

```python
houses.head()
```

```python
houses.describe()
```

```python
#Removing columns with no impact on Price
houses.dropna(subset=['AYB','Y','QUADRANT','X','ASSESSMENT_NBHD','WARD','ZIPCODE','LATITUDE','LONGITUDE','CENSUS_TRACT'], inplace=True)
```

```python
#gross building area (GBA) and living gorss buidling area (living_gba)
#fILLING NAN Values
mean_livingGBA = np.mean(houses.LIVING_GBA)
houses['LIVING_GBA']=houses.LIVING_GBA.fillna(mean_livingGBA)
mean_GBA = np.mean(houses.GBA)
houses['GBA']=houses.GBA.fillna(mean_GBA)
```

```python
houses['SALEDATE'] = pd.to_datetime(houses['SALEDATE'])
houses.dropna(subset=['SALEDATE'], inplace=True)
#GIS_LAST_MOD_DTTM
houses['GIS_LAST_MOD_DTTM'] = pd.to_datetime(houses['GIS_LAST_MOD_DTTM'])
houses.dropna(subset=['GIS_LAST_MOD_DTTM'], inplace=True)
```

```python
houses["ZIPCODE"] = houses.ZIPCODE.astype(object)
houses["ASSESSMENT_SUBNBHD"] = houses.ASSESSMENT_SUBNBHD.astype(object)
```

```python
houses["NUM_UNITS"] = houses["NUM_UNITS"].fillna(0)
houses["STORIES"] = houses["STORIES"].fillna(0)
houses["KITCHENS"] = houses["KITCHENS"].fillna(0)
houses["GRADE"] = houses["GRADE"].fillna("None")
houses["STRUCT"] = houses["STRUCT"].fillna("None")
houses["ROOF"] = houses["ROOF"].fillna("None")
houses["STYLE"] = houses["STYLE"].fillna("None")
houses["CNDTN"] = houses["CNDTN"].fillna("None")
```

```python
#removing columns with no impact on price
houses = houses.drop(['CMPLX_NUM', 'YR_RMDL','FULLADDRESS','NATIONALGRID','CENSUS_BLOCK',
                             'STATE','CITY','EXTWALL','INTWALL'],axis=1)
```

```python
houses = houses.drop_duplicates()
```

```python
nonPrice = houses[houses['PRICE'].isnull()]
nonPrice.head()
```

```python
houses.dropna(subset=['PRICE'], inplace=True)
```

```python
int_cols = [key for key in dict(houses.dtypes) if dict(houses.dtypes)[key] in ['int64','float64']]
int_cols
```

```python
#dropping housing outliers
def drop_outliers(df, field_name):
    distance = 1.5 * (np.percentile(df[field_name], 75) - np.percentile(df[field_name], 25))
    df.drop(df[df[field_name] > distance + np.percentile(df[field_name], 75)].index, inplace=True)
    df.drop(df[df[field_name] < np.percentile(df[field_name], 25) - distance].index, inplace=True)
```

```python
for i in int_cols:
    drop_outliers(houses,i)
```

```python
sns.boxplot(y=houses.PRICE)
```

```python
housing_prices = houses[['PRICE', 'WARD', 'SALEDATE']]
```

```python
display(housing_prices)
```

```python
ward_dfs = {}
for ward, group in housing_prices.groupby('WARD'):
    # Sort the DataFrame by 'SALEDATE' in ascending order
    sorted_df = group.drop(columns=['WARD']).sort_values(by='SALEDATE', ascending=True)
    # Set 'SALEDATE' as the index
    sorted_df.set_index('SALEDATE', inplace=True)
    # Store the DataFrame in the ward_dfs dictionary with ward name as key
    ward_dfs[ward] = sorted_df

# Access individual DataFrames by ward name
ward_1_df = ward_dfs.get('Ward 1', pd.DataFrame())  # Provide an empty DataFrame if 'Ward 1' doesn't exist
ward_2_df = ward_dfs.get('Ward 2', pd.DataFrame())  # Provide an empty DataFrame if 'Ward 2' doesn't exist
ward_3_df = ward_dfs.get('Ward 3', pd.DataFrame())  # Provide an empty DataFrame if 'Ward 3' doesn't exist
ward_4_df = ward_dfs.get('Ward 4', pd.DataFrame())  # Provide an empty DataFrame if 'Ward 4' doesn't exist
ward_5_df = ward_dfs.get('Ward 5', pd.DataFrame())  # Provide an empty DataFrame if 'Ward 5' doesn't exist
ward_6_df = ward_dfs.get('Ward 6', pd.DataFrame())  # Provide an empty DataFrame if 'Ward 6' doesn't exist
ward_7_df = ward_dfs.get('Ward 7', pd.DataFrame())  # Provide an empty DataFrame if 'Ward 7' doesn't exist
ward_8_df = ward_dfs.get('Ward 8', pd.DataFrame())  # Provide an empty DataFrame if 'Ward 8' doesn't exist
```

```python
import matplotlib.pyplot as plt
plt.style.use('default')
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize=(16,8), dpi = 150)

ward_1_df['PRICE'].plot(label='ward 1', color = 'orange')
ward_2_df['PRICE'].plot(label='ward 2', color = 'blue')
ward_3_df['PRICE'].plot(label='ward 3', color = 'green')
ward_4_df['PRICE'].plot(label='ward 4', color = 'red')
ward_5_df['PRICE'].plot(label='ward 5', color = 'purple')
ward_6_df['PRICE'].plot(label='ward 6', color = 'brown')
ward_7_df['PRICE'].plot(label='ward 7', color = 'pink')
ward_8_df['PRICE'].plot(label='ward 8', color = 'gray')

plt.title('Housing Prices in Washington DC')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.show()


```
